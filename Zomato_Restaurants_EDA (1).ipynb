```python
import pandas as pd
import numpy as np

```

```python
# Load the dataset
data = pd.read_csv('/zomato_restaurants_in_India.csv')

# Display the first few rows of the dataframe to understand its structure and the type of data it contains
data.head()
```

```python
# Display the first 10 rows of the dataset in tabular format for a clearer view
data.head(10)
```

```python
# Check for missing values in the dataset
missing_data = data.isnull().sum()

# Check data types for consistency
data_types = data.dtypes

missing_data, data_types
```

```python
# Drop the 'zipcode' column due to a large number of missing values
df = data.drop(columns='zipcode')

# Fill missing values for 'cuisines' and 'timings' with "Not Available"
df['cuisines'] = df['cuisines'].fillna('Not Available')
df['timings'] = df['timings'].fillna('Not Available')

# For 'address' and 'opentable_support', let's fill missing values with a placeholder or a default value
df['address'] = df['address'].fillna('Address not available')
df['opentable_support'] = df['opentable_support'].fillna(0)  # Assuming 0 as 'No support'

# Check the dataset after handling missing values
df.isnull().sum()
```

```python
# Display the first 10 rows in tabular format
df.head(10)
```

```python
# Convert columns to 'category' data type
categorical_columns = ['establishment', 'cuisines', 'rating_text', 'highlights']
df[categorical_columns] = df[categorical_columns].astype('category')

# Convert 'opentable_support', 'delivery', and 'takeaway' to boolean
# Here, assuming that -1 represents 'False' and other positive values represent 'True'
boolean_columns = ['opentable_support', 'delivery', 'takeaway']
df[boolean_columns] = df[boolean_columns].replace(-1, False).replace(1, True).astype(bool)

# Display data types after conversion
df.dtypes
```

```python
# Remove [" and "] from the 'establishment' column and convert to just text
df['establishment'] = df['establishment'].str.replace(r"[\['\]]", '', regex=True)

# Display the first few rows to confirm the change
df[['establishment']].head()
```

```python
# Generate summary statistics for numerical features in the dataset
numerical_summary = df.describe()

numerical_summary
```

```python
# Check for duplicate rows in the dataset
duplicate_count = df.duplicated().sum()

duplicate_count
```

```python
# Find duplicate rows, keeping the first occurrence as unique
duplicates = df[df.duplicated(keep=False)]

# Get a sample of 10 duplicate rows for examination
sample_duplicates = duplicates.sample(n=10, random_state=1)

# Analyze duplicates for patterns, focusing on cities, establishments, and cuisines
duplicate_analysis = {
    'Cities': duplicates['city'].value_counts(),
    'Establishments': duplicates['establishment'].value_counts(),
    'Cuisines': duplicates['cuisines'].value_counts()
}

sample_duplicates, {key: value.head() for key, value in duplicate_analysis.items()}
```

```python
# Remove exact duplicates from the dataset
zomato_data_no_duplicates = df.drop_duplicates()

# Count the remaining data
remaining_count = zomato_data_no_duplicates.shape[0]

# Find similar but not exact duplicates (for example, by comparing names and addresses)
similar_duplicates = zomato_data_no_duplicates.duplicated(subset=['name', 'address', 'city'], keep=False)
similar_count = similar_duplicates.sum()

# Calculate unique entries
unique_count = remaining_count - similar_count

remaining_count, similar_count, unique_count
```

```python


# Function to calculate IQR and identify outliers
def find_outliers(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - (1.5 * IQR)
    upper_bound = Q3 + (1.5 * IQR)

    outliers = df[(df[column] < lower_bound) | (df[column] > upper_bound)]
    return outliers

# Analyzing outliers for key numerical columns
key_columns = ['average_cost_for_two', 'aggregate_rating', 'votes', 'photo_count']
outliers_summary = {}

for column in key_columns:
    outliers = find_outliers(zomato_data_no_duplicates, column)
    outliers_summary[column] = {
        'Outlier Count': outliers.shape[0],
        'Percentage': (outliers.shape[0] / zomato_data_no_duplicates.shape[0]) * 100
    }

outliers_summary
```

```python
# Function to cap outliers based on the IQR method
def cap_outliers(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - (1.5 * IQR)
    upper_bound = Q3 + (1.5 * IQR)

    # Capping the data
    df[column] = np.where(df[column] < lower_bound, lower_bound, df[column])
    df[column] = np.where(df[column] > upper_bound, upper_bound, df[column])

    return df

# Capping outliers for key numerical columns
for column in key_columns:
    zomato_data_no_duplicates = cap_outliers(zomato_data_no_duplicates, column)

# Display the first few rows of the updated dataset
zomato_data_no_duplicates.head()
```
